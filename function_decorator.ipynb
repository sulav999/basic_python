```python
def call_twice(function):
    def wrapper(*args, **kwargs):
        function(*args, **kwargs)
        function(*args, **kwargs)
    return wrapper
```

```python
@call_twice
def greet(name):
    print(f"hello {name}")
```

```python
greet("john")
```

```python
@call_twice
def sum_num(a,b):
    print(a+b)
    
```

```python
sum_num(2,3)
```

```python
# 
def pin_check(func):
    def wrapper(*args,**kwargs):
        if kwargs.get('pin')==1234:
            return func(*args,**kwargs)
        else:
            print("invalid pin")
    return wrapper
    
```

```python
@pin_check
def withdraw(amt, pin):
    print(f"{amt} withdraw")
```

```python
withdraw(12,pin=1234)
```

```python
#sending parameter in @Decerators


def call_n_times(n=1):
    def decurator(function):
        def wrapper(*args, **kwargs):
            for i in range(n):
                function(*args,**kwargs)
        return wrapper
    return decurator
```

```python
@call_n_times(n=3)
def greet(name):
    print(f"hello {name}")
```

```python
greet('john')
```

```python
def sum_rec(list1):
    if len(list1)==1:
        return list1[0]
    else:
        return list1[0]+sum_rec(list1[1:])
```

```python
sum_rec([1,2,3,4])
```

```python
#factorial
#SCALla(data science)

def factorial(n):
    if n==1 or n==0:
        return 1
    else:
        return n * factorial(n-1)
```

```python
factorial(5)
```

```python
#febonancis series

def fibo(n):
    if n==1 or n==2:
        return 1
    else:
        return fibo(n-1) + fibo(n-2)
```

```python
fibo(6)
```

```python
#[1,2[3,4]5] sum using recursion

def rec_sum_list(list_1):
    sum_list=0
    for element in list_1:
        if type(element)==type([]):  
            sum_list = sum_list + rec_sum_list(element)
        else:
            sum_list = sum_list + element
    return sum_list
```

```python
rec_sum_list([1,2,[3,4],5])
```

```python
#generator (have defult libary for generator in python called--> itertools)#list comprehansion
#only execute for one time 

def gen_list(upto):
    for i in range(upto):
        yield i
```
